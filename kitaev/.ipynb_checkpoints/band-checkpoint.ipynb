```python

def build_zero_flux_standard_u_configuration(N1, N2, bc1, bc2): # bc1 = 1 代表 a1 方向 PBC, bc1 = -1 代表 a1 方向 APBC
    """
    
    
    Parameters
    ----------
    N1, N2 : int
        N1, N2 分别代表 a1, a2 方向上 unit cell 数量

    bc1, bc2 : int
        bc1 = 1 代表 a1 方向 PBC, 也即 a1 方向上跨边界的 u_jk 取 +1
        bc1 = -1 代表 a1 方向 APBC, 也即 a1 方向上跨边界的 u_jk 取 -1
    
    Returns
    -------
    u : numpy.ndarray, shape : (2*N1*N2, 2*N1*N2)
        用来保存 u 构型

    Notes
    -----
    
    """
    
    N = N1 * N2 # N1 为 a_1 方向 unit cell 数量, N2 为 a_2 方向 unit cell 数量
    M = np.zeros((2 * N, 2 * N)) # M 储存 u_jk(j 属于 A 子格，k 属于 B 子格)构型
    u = np.zeros((2 * N, 2 * N)) 
    
    def index(n1, n2, sub):
        return n1 + n2 * N1 + N1 * N2 * sub # A子格对应 sub=0,B子格对应 sub=1
    
    for n2 in range(N2):
        for n1 in range(N1):
            j = index(n1, n2, 0) # j 为(n1, n2)unit cell 内 A 子格格点的单指标
            
            kx = index(n1, n2, 1) # kx 为(n1, n2)unit cell 内 B 子格格点的单指标
            M[j, kx] = 1 # j 与 kx 由 x-bond 相连，一定不跨越边界
            
            ky = index((n1 + 1) % N1, n2, 1) # ky 为 (n1, n2) 往 a1 方向走一步得到的 unit cell 内 B 子格格点的单指标j 与 ky 由 y-bond 相连
            if n1 == N1 - 1:
                M[j, ky] = bc1
            else:
                M[j, ky] = 1

            kz = index(n1, (n2 + 1) % N2, 1) # kz 为 (n1, n2) 往 a2 方向走一步得到的 unit cell 内 B 子格格点的单指标，j 与 kz 由 z-bond 相连
            if n2 == N2 - 1:
                M[j, kz] = bc2
            else:
                M[j, kz] = 1
      
    u = M - M.T                        
    return u
    
############## test u_std ##############

# u_test = build_zero_flux_standard_u_configuration(2, 2, -1, -1)
# print(u_test)
```

```python
def build_t_matrix(N1, N2, Kx, Ky, Kz, u):
    N = N1 * N2
    t = np.zeros((N, N))
    K = np.array([Kx, Ky, Kz])
    for i in range(N):
        for j in range(N):
            sum = 0
            r_i_vec = np.array([i % N1, i // N1])
            r_j_vec = np.array([j % N1, j // N1])
            
            for delta_vec in np.array([[0, 0], [1, 0], [0, 1]]):
                tmp1 = (r_j_vec[0] - delta_vec[0] + N1) % N1
                tmp2 = (r_j_vec[1] - delta_vec[1] + N2) % N2
                if (np.array_equal(r_i_vec, np.array([tmp1, tmp2]))):
                    delta_func = 1
                else:
                    delta_func = 0
                    
                tmp3 = delta_vec[0] + delta_vec[1] * 2
                K_delta = K[tmp3]

                sum += delta_func * K_delta
                
            sum *= u[i, j + N]
            t[i, j] = sum
    return t
```

```python
def flip_single_bond(N1, N2, n1, n2, sub, bond, u):
    def index(n1, n2, sub):
        return n1 + n2 * N1 + N1 * N2 * sub # A子格对应 sub=0,B子格对应 sub=1
    j = index(n1, n2, sub)
    if sub == 0:
        if bond == 1: # x-bond
            k = index(n1, n2, 1)
        elif bond == 2: # y-bond
            k = index((n1+1)%N1, n2, 1)
        elif bond == 3: # z-bond
            k = index(n1, (n2+1)%N2, 1)
    if sub == 1:
        if bond == 1: # x-bond
            k = index((n1-1+N1)%N1 , n2, 0)
        elif bond == 2: # y-bond
            k = index(n1, n2, 0)
        elif bond == 3: # z-bond
            k = index(n1, (n2-1+N2)%N2, 0)       

    u[j, k] = -u[j, k]
    u[k, j] = -u[k, j]
    return u
```

```python
def build_BdG_Hamiltanian(N1, N2, Kx, Ky, Kz, u):
    def build_t_matrix(N1, N2, Kx, Ky, Kz, u):
        N = N1 * N2
        t = np.zeros((N, N))
        K = np.array([Kx, Ky, Kz])
        for i in range(N):
            for j in range(N):
                sum = 0
                r_i_vec = np.array([i % N1, i // N1])
                r_j_vec = np.array([j % N1, j // N1])
                
                for delta_vec in np.array([[0, 0], [1, 0], [0, 1]]):
                    tmp1 = (r_j_vec[0] - delta_vec[0] + N1) % N1
                    tmp2 = (r_j_vec[1] - delta_vec[1] + N2) % N2
                    if (np.array_equal(r_i_vec, np.array([tmp1, tmp2]))):
                        delta_func = 1
                    else:
                        delta_func = 0
                        
                    tmp3 = delta_vec[0] + delta_vec[1] * 2
                    K_delta = K[tmp3]
        
                    sum += delta_func * K_delta
                    
                sum *= u[i, j + N]
                t[i, j] = sum
        return t
    t = build_t_matrix(N1, N2, Kx, Ky, Kz, u)
    Xi = t + t.T
    Delta = -(t - t.T)
    H = np.block([
        [Xi, Delta],
        [-Delta, -Xi.T]
    ])
    return H
```

```python
def diagonalize_BdG_Hamiltonian(H, N): # H 是 2N 维方阵
    eigvals, eigvecs = np.linalg.eigh(H)
    positive_sorted_eigvals = eigvals[N:] # numpy默认本征值升序，正的本征值在后半部分
    W = eigvecs[:N, N:]
    V = eigvecs[N:, N:]
    U = np.block([
        [W, V.conj()],
        [V, W.conj()]
    ])
    
    return positive_sorted_eigvals, U
```

```python
def build_U0_prime_matrix(U0, N):
    I_N = np.eye(N, dtype=complex)
    U0_prime = np.block([
        [I_N, I_N],
        [-1j*I_N, 1j*I_N]
    ]) @ U0
    U0_prime_11 = U0_prime[:N, :N]
    U0_prime_12 = U0_prime[:N, N:]
    U0_prime_21 = U0_prime[N:, :N]
    U0_prime_22 = U0_prime[N:, N:]
    return U0_prime_11, U0_prime_12, U0_prime_21, U0_prime_22
```

```python
def compute_hopping_amplitude(N1, N2, Kx, Ky, Kz, n1, n2, bc1, bc2):
    sorted_positive_eigvals = []
    GS_energy = []
    N = N1 * N2
    u_std = build_zero_flux_standard_u_configuration(N1, N2, bc1, bc2)
    H_std = build_BdG_Hamiltanian(N1, N2, Kx, Ky, Kz, u_std)
    sorted_positive_eigvals_std, U0 = diagonalize_BdG_Hamiltonian(H_std, N)
    sorted_positive_eigvals.append(sorted_positive_eigvals_std)
    GS_energy_std = -np.sum(sorted_positive_eigvals_std)/2
    GS_energy.append(GS_energy_std)
    # print(f"GS_energy:{GS_energy_std}")
    U0_prime_11, U0_prime_12, U0_prime_21, U0_prime_22 = build_U0_prime_matrix(U0, N)
    U0_dagger = U0.T.conj()
    W_tilde_configs = [
        [None, None, None, None],
        [None, None, None, None]
    ]
    V_tilde_configs = [
        [None, None, None, None],
        [None, None, None, None]
    ]
    F_tilde_configs = [
        [None, None, None, None],
        [None, None, None, None]
    ]
    hopping = []
    def build_T_1_matrix(N1, N2):
        single_block = np.eye(N1, k=-1)
        single_block[0, N1-1] = -1
        T_1 = np.kron(np.eye(N2), single_block)
        return T_1
    def build_T_2_matrix(N1, N2):
        single_block = np.eye(N2, k=-1)
        single_block[0, N2-1] = -1
        T_2 = np.kron(single_block, np.eye(N1))
        return T_2
    T_1 = build_T_1_matrix(N1, N2)
    T_2 = build_T_2_matrix(N1, N2)
    
    for flipped_bond in [1, 2, 3]:
        u_tmp_A = flip_single_bond(N1, N2, n1, n2, 0, flipped_bond, u_std.copy()) # A子格
        H_tmp_A = build_BdG_Hamiltanian(N1, N2, Kx, Ky, Kz, u_tmp_A)
        sorted_positive_eigvals_tmp_A, U_tmp_A = diagonalize_BdG_Hamiltonian(H_tmp_A, N)
        sorted_positive_eigvals.append(sorted_positive_eigvals_tmp_A)
        GS_energy_tmp = -np.sum(sorted_positive_eigvals_tmp_A) / 2
        GS_energy.append(GS_energy_tmp)
        # print(sorted_eigvals_tmp_A)
        print(f"GS_energy:{GS_energy_tmp}")
        if flipped_bond == 1:
            U_tmp_B = np.kron(np.eye(2), T_2) @ U_tmp_A
        elif flipped_bond == 2:
            U_tmp_B = np.kron(np.eye(2), T_2) @ np.kron(np.eye(2), T_1.T) @ U_tmp_A
        elif flipped_bond == 3:
            U_tmp_B = U_tmp_A
        U_tilde_A = U0_dagger @ U_tmp_A
        U_tilde_B = U0_dagger @ U_tmp_B
        
        W_tilde_A = U_tilde_A[:N, :N]
        V_tilde_A = U_tilde_A[N:, :N]
        F_tilde_A = V_tilde_A.conj() @ np.linalg.inv(W_tilde_A.conj())
        W_tilde_B = U_tilde_B[:N, :N]
        V_tilde_B = U_tilde_B[N:, :N]
        F_tilde_B = V_tilde_B.conj() @ np.linalg.inv(W_tilde_B.conj())
        W_tilde_configs[0][flipped_bond] = W_tilde_A
        V_tilde_configs[0][flipped_bond] = V_tilde_A
        F_tilde_configs[0][flipped_bond] = F_tilde_A
        W_tilde_configs[1][flipped_bond] = W_tilde_B
        V_tilde_configs[1][flipped_bond] = V_tilde_B
        F_tilde_configs[1][flipped_bond] = F_tilde_B
    
    Ex = np.diag(GS_energy[1] + sorted_positive_eigvals[1] - GS_energy[0])
    Ey = np.diag(GS_energy[2] + sorted_positive_eigvals[2] - GS_energy[0])
    Ez = np.diag(GS_energy[3] + sorted_positive_eigvals[3] - GS_energy[0])
    
    all_bonds = {1, 2, 3}
    for sub in [0, 1]:
        for bond_1, bond_2 in [[1, 2], [2, 3], [3, 1]]:
            bond_left = (all_bonds - {bond_1, bond_2}).pop() #比如 bond_1=2,bond_2=3,则bond_left=1
            if sub == 0:
                i = N1 * n2 + n1 #i对应r\in A的unit cell单指标
                if bond_left == 1:
                    j = N1 * n2 + n1
                elif bond_left == 2:
                    j = N1 * n2 + (n1 + 1) % N1
                elif bond_left == 3:
                    j = N1 * ((n2+1) % N2) + n1
            if sub == 1:
                if bond_left == 1: # y-z
                    i = N1 * ((n2+1) % N2) + n1
                    j = N1 * ((n2+1) % N2) + n1
                elif bond_left == 2: # z-x
                    i = N1 * ((n2+1) % N2) + (n1-1+N1) % N1
                    j = N1 * ((n2+1) % N2) + n1
                elif bond_left == 3: # x-y
                    i = N1 * n2 + n1
                    j = N1 * ((n2+1) % N2) + n1
                    
            N_1_tilde = np.sqrt(np.abs(np.linalg.det(W_tilde_configs[sub][bond_1])))
            N_2_tilde = np.sqrt(np.abs(np.linalg.det(W_tilde_configs[sub][bond_2])))
            M = np.block([
                [F_tilde_configs[sub][bond_2], -np.eye(N)],
                [np.eye(N), -F_tilde_configs[sub][bond_1].conj()]
            ])
            overlap = N_1_tilde * N_2_tilde * (-1)**(N*(N+1)/2) * pfaffian.pfaffian(M)
            print(overlap)
            middle_matrix = -np.linalg.inv(M) + np.block([ [np.zeros((N, N)), np.eye(N)], [np.zeros((N, N)), np.zeros((N, N))] ])
            left_WV = np.hstack([V_tilde_configs[sub][bond_1].T.conj(), W_tilde_configs[sub][bond_1].T.conj()])
            right_WV = np.vstack([W_tilde_configs[sub][bond_2], V_tilde_configs[sub][bond_2]])
            cor_alpha_1_alpha_2_dag = left_WV @ middle_matrix @ right_WV
            cor_alpha_1_c_i_A = left_WV @ middle_matrix @ np.vstack([U0_prime_11[i, :].conj().reshape(-1, 1) ,U0_prime_12[i, :].conj().reshape(-1, 1)])
            cor_alpha_1_c_j_B = left_WV @ middle_matrix @ np.vstack([U0_prime_21[j, :].conj().reshape(-1, 1) ,U0_prime_22[j, :].conj().reshape(-1, 1)])
            cor_c_i_A_alpha_2_dag = np.hstack([U0_prime_12[i,:].reshape(1,-1), U0_prime_11[i,:].reshape(1,-1) ]) @ middle_matrix @ right_WV
            cor_c_j_B_alpha_2_dag = np.hstack([U0_prime_22[j,:].reshape(1,-1), U0_prime_21[j,:].reshape(1,-1) ]) @ middle_matrix @ right_WV
            cor_c_i_A_c_j_B = np.hstack([U0_prime_12[i,:].reshape(1,-1), U0_prime_11[i,:].reshape(1,-1) ]) @ middle_matrix @ np.vstack(
                [U0_prime_21[j, :].conj().reshape(-1, 1) ,U0_prime_22[j, :].conj().reshape(-1, 1)]
            ) # 还是个1\times 1二维数组
            
            cor_alpha_1_c_i_A_c_j_B_alpha_2_dag = cor_alpha_1_c_i_A @ cor_c_j_B_alpha_2_dag - cor_alpha_1_c_j_B @ cor_c_i_A_alpha_2_dag + \
            cor_alpha_1_alpha_2_dag * cor_c_i_A_c_j_B[0, 0]
            hopping_amp = h_mag[bond_left-1] * overlap * (1j * cor_alpha_1_alpha_2_dag - cor_alpha_1_c_i_A_c_j_B_alpha_2_dag)
            hopping.append(hopping_amp)
    return hopping[0], hopping[1], hopping[2], hopping[3], hopping[4], hopping[5], Ex, Ey, Ez # T^A_xy, T^A_yz, T^A_zx, T^B_xy, T^B_yz, T^B_zx
    
```

```python
N1, N2, Kx, Ky, Kz, kappa = 10, 10, 1, 1, 1, 0.1
h_mag = np.array([0.1, 0.1, 0.1]) # 磁场
N = N1 * N2
n1, n2 = N1//2, N2//2 
bc1, bc2 = -1, -1
a1 = np.array([1.0, 0.0])
a2 = np.array([0.5, np.sqrt(3)/2])
b1 = np.array([2*np.pi, -2*np.pi/np.sqrt(3)])
b2 = np.array([0.0, 4*np.pi/np.sqrt(3)])
# print(b1, b2)
```

```python
T_A_xy, T_A_yz, T_A_zx, T_B_xy, T_B_yz, T_B_zx, Ex, Ey, Ez = compute_hopping_amplitude(N1, N2, Kx, Ky, Kz, n1, n2, bc1, bc2)
```

```python
def k_mesh(N1, N2, bc1=-1, bc2=-1):
    factor1 = 0.0 if bc1 == 1 else 0.5
    factor2 = 0.0 if bc2 == 1 else 0.5

    k_list = []
    for n1 in range(N1):
        for n2 in range(N2):
            k1 = (n1 + factor1) / N1
            k2 = (n2 + factor2) / N2
            k = k1 * b1 + k2 * b2
            k_list.append((k, k1, k2))
    return k_list
```

```python
k_list = k_mesh(N1, N2, bc1=-1, bc2=-1)
```

```python
Gamma = np.array([0.0, 0.0])
K = np.array([4*np.pi/3, 0.0])
M = np.array([np.pi, np.pi/np.sqrt(3)])

def k_path(points, n=100):
    k_list = []
    k_dist = []
    for i in range(len(points)-1):
        seg = np.linspace(points[i], points[i+1], n, endpoint=False)
        for k in seg:
            k_list.append(k)
            if len(k_list) > 1:
                k_dist.append(k_dist[-1] + np.linalg.norm(k_list[-1]-k_list[-2]))
            else:
                k_dist.append(0.0)
    k_list.append(points[-1])
    k_dist.append(k_dist[-1] + np.linalg.norm(points[-1]-k_list[-2]))
    return np.array(k_list), np.array(k_dist)

k_list, k_dist = k_path([Gamma, K, M, Gamma])
```

```python
def build_h_k(k, T_A_xy, T_A_yz, T_A_zx, T_B_xy, T_B_yz, T_B_zx, Ex, Ey, Ez, a1, a2):
    phase1 = np.exp(-1j * k @ a1)
    phase2 = np.exp( 1j * k @ a2)
    phase3 = np.exp( 1j * k @ (a1 - a2))

    h_k = np.block([
        [Ex,                T_A_xy + phase1*T_B_xy,       (T_A_zx + phase2*T_B_zx).conj().T],
        [(T_A_xy + phase1*T_B_xy).conj().T, Ey,            T_A_yz + phase3*T_B_yz],
        [T_A_zx + phase2*T_B_zx, (T_A_yz + phase3*T_B_yz).conj().T, Ez]
    ])
    return h_k
```

```python
bands = []
for k in k_list:
    h_k = build_h_k(k, T_A_xy, T_A_yz, T_A_zx, T_B_xy, T_B_yz, T_B_zx, Ex, Ey, Ez, a1, a2)
    eigvals = np.linalg.eigh(h_k)[0]
    bands.append(eigvals)

bands = np.array(bands)

for n in range(3*N):
    plt.plot(k_dist, bands[:, n], color='black', linewidth=0.6)

plt.xticks(
    [k_dist[0], k_dist[100], k_dist[200], k_dist[300]],
    [r'$\Gamma$', r'$K$', r'$M$', r'$\Gamma$']
)
plt.ylabel("Energy")
plt.xlabel(r"$\mathbf{k}$ distance")
plt.title(f"Band Structure for $N_1,N_2={N1,N2}$,Kx,Ky,Kz={Kx,Ky,Kz},h_mag={h_mag}", fontsize=10)
plt.grid(alpha=0.3)
plt.savefig("band_structure.png", dpi=300, bbox_inches="tight")
plt.show()
```
